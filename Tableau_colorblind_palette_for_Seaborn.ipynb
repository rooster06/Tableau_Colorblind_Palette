```python
import numpy as np
import pandas as pd
from scipy.stats import norm

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## create a dummy dataset

df = pd.DataFrame(data = np.random.random(size=(10,10))
                  , columns = ['A','B', 'C' ,'D','E','F','G','H','I','J'])

df.head()
```

 ## color codes from RGB values in python
 - color codes can be created from rgb values using the following commands in python
 
> '#%02x%02x%02x' % (red_value ,green_value,  blue_value) 

- values must be between 0 to 255

```python
## example
'#%02x%02x%02x' % (189,59 ,49)
```

## Tableau Colorblind palette color codes
 
Values obtained from 
http://tableaufriction.blogspot.com/2012/11/finally-you-can-use-tableau-data-colors.html

copy paste this list into your code and use for tableaus colorblind palette

```python

tableauColorBlind = ["#006ba4", "#ff800e"
                     , "#ababab", "#595959"
                     , "#5f9ed1", "#c85200"
                     ,"#898989","#a2c8ec"
                     ,"#ffbc79","#cfcfcf"]
```

### Seaborn default colorblind palette V tableau colorblind palette

```python
sns.palplot(sns.color_palette("colorblind", 10))
```

```python
sns.palplot(sns.color_palette(tableauColorBlind, 10))
```

## Examples

```python
sns.set(rc={'figure.figsize':(32,20)})
## set palette
sns.set_palette(sns.color_palette("colorblind", 10))
sns.boxplot(x="variable", y="value", data=pd.melt(df))
```

```python
sns.set(rc={'figure.figsize':(32,20)})
## set palette
sns.set_palette(tableauColorBlind)
sns.boxplot(x="variable", y="value", data=pd.melt(df))
```
